```python
import os
import urllib.request
```

```python
url = "https://opendata.chmi.cz/meteorology/climate/historical_csv/data/daily/temperature/dly-0-203-0-11514-T.csv"
output_dir = "data_raw"

# Extract the filename from the end of the URL
filename = "klementinum_teplota_" + url.split("/")[-1]
file_path = os.path.join(output_dir, filename)

# Create the 'data_raw' directory if it doesn't already exist
os.makedirs(output_dir, exist_ok=True)

print(f"Downloading {filename}...")

# Download and save the file
try:
    urllib.request.urlretrieve(url, file_path)
    print(f"Success! File saved to: {file_path}")
except Exception as e:
    print(f"An error occurred during download: {e}")
```
