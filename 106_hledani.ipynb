```python
import os
import polars as pl
import altair as alt
```

```python
csvs = [x for x in os.listdir('data_raw') if 'time_series_CZ_' in x]

df = pl.DataFrame()
for csv in csvs:
    df = pl.concat(
        [
            df,
            pl.read_csv(os.path.join("data_raw",csv)).unpivot(index='Time').with_columns(pl.col('Time').str.to_date())
        ]
    )
```

```python
df
```

```python
df.group_by(pl.col('Time').dt.month()).agg(pl.col('value').median()).sort(by='Time')
```

```python
alt.Chart(
    df.filter(pl.col('variable') == "jak začít běhat").group_by(pl.col('Time').dt.month()).agg(pl.col('value').median()).sort(by='Time')
).mark_bar(
).encode(
    alt.X('Time:N'),
    alt.Y('value:Q')
)
```
