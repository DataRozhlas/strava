```python
import polars as pl
```

```python
df1 = pl.read_json('data/korelace_behani.json').with_columns(pl.lit("běh").alias("sport")).rename({'korelace s počtem běhů':'korelace s počtem výkonů'})
df2 = pl.read_json('data/korelace_cyklistika.json').with_columns(pl.lit("cyklistika").alias("sport")).rename({'korelace s počtem jízd':'korelace s počtem výkonů'})
df_korelace = pl.concat([df1, df2])
df_korelace
```

```python
df_korelace.write_json("data/graf_korelace.json")
```

```python
df1 = pl.read_json('data/graf_rok_behani.json').with_columns(pl.lit("běh").alias("sport"))
df2 = pl.read_json('data/graf_rok_cyklo.json').with_columns(pl.lit("cyklistika").alias("sport"))
df_pokusy = pl.concat([df1, df2])
df_pokusy
```

```python
df_pokusy.write_json("data/graf_pokusy.json")
```
