```python
import polars as pl
```

```python
df = pl.read_parquet("data_raw/segments/*.parquet")

df = df.with_columns(
    (pl.col("date").cast(pl.Int64) * 1000000).cast(pl.Datetime(time_zone='CET'))
)
df = df.filter(pl.col('date').dt.hour() > 22)
df = df.sort(by="date")
```

```python
df.columns
```

```python
df.group_by_dynamic("date", by="name", every="1d").agg(pl.col("effort_count").max())
```

```python
df.filter(pl.col('country').str.contains('Cze')).filter(pl.col('activity_type') == 'Run').group_by("name").agg(pl.col('effort_count').cast(int).max()).sort(by='effort_count',descending=True).head(n=10)
```

```python
import altair as alt
```

```python
segm = df.filter(pl.col('name') == 'Vojenský ovál 400M').group_by_dynamic("date", every="1d").agg(pl.col("effort_count").max())
segm = segm.with_columns(pl.col("effort_count").diff().alias("daily_difference")).drop_nulls()
segm = segm.with_columns(pl.col("date").cast(str))
alt.Chart(segm).mark_bar().encode(
    x='date:O',
    y='daily_difference:Q'
)
```
